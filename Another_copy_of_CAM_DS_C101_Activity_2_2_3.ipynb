```python
# Import libraries
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt #Visualisation
import seaborn as sns #Visualisation

data = pd.read_csv("https://raw.githubusercontent.com/fourthrevlxd/cam_dsb/main/insurance_updated.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
# Create a scatter plot
sns.scatterplot(data=data, x="number_of_children", y="charges")
plt.show()
```

```python
# Create a scatter plot
sns.scatterplot(data=data, x="bmi", y="charges")
plt.show()
```

```python
from scipy.stats import pearsonr

# data
x = data.loc[:, "number_of_children"]
y = data.loc[:, "charges"]

# Calculate Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(x, y)

print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)
```

Interpretation of strength and direction:

The correlation coefficient is close to zero (0.031). This means there is no meaningful linear relationship between number of children and insurance charges.

Statistical significance:

The p-value (0.332) is much greater than 0.05, so the result is not statistically significant.



```python
from scipy.stats import pearsonr

# data
x = data.loc[:, "bmi"]
y = data.loc[:, "charges"]

# Calculate Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(x, y)

print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)
```

Interpretation of relationship:

The coefficient is positive (0.709). As BMI increases, insurance charges also increase.

Statistical significance:

The p-value is extremely small (much smaller than 0.05). This means the probability of observing this correlation being random is almost zero.

Therefore, the correlation is statistically significant.

```python
from scipy.stats import pearsonr

# data
x = data.loc[:, "age"]
y = data.loc[:, "charges"]

# Calculate Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(x, y)

print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)
```

Interpretation of this relationship.

Correlation coefficient is 0.08 which points to a very weak positive relationship. This means insurance charges tend to rise only slightly with age.

Statistical significance:

Since the p-value (â‰ˆ0.012) is below the 0.05 threshold, the relationship is statistically significant.

So the weak correlation is unlikely to be due to chance, but it is still a very small effect compared with BMI.

Limitations and Bias
It is essential to present these findings with the following limitations in mind:

Correlation does not imply causation: While a strong correlation exists, it does not prove that a high BMI directly causes high insurance costs. There may be other unobserved variables such as smoking status, chronic conditions  that influences both BMI and insurance charges. Without this additional data, a direct causal link cannot be established.


Linearity assumption: Pearson correlation assumes a linear relationship. If the relationship is non-linear (e.g., costs rise sharply only above a certain BMI threshold), the correlation may underestimate the true association.

Sensitivity to outliers: A few extreme values in charges (very high claims) can inflate or distort the correlation coefficient.


In conclusion, the analysis provides robust statistical evidence to support a tiered insurance program. However, it's vital to frame this evidence correctly, acknowledging that it identifies strong correlations, not absolute causation, and is limited by the variables available in the dataset.
